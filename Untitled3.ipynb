```python
def cargar_dataset(ruta_csv):
    """
    Lee el archivo CSV y extrae los textos y sus clases correspondientes.

    Por qué se hace así:
        El enunciado prohíbe usar pandas u otras librerías no autorizadas,
        por lo que se usa el módulo estándar `csv` de Python. Se itera
        fila por fila para construir dos listas paralelas: una con los
        textos y otra con las etiquetas de clase. Esto facilita el acceso
        por índice en las etapas posteriores.

    Parámetros:
        ruta_csv (str): Ruta al archivo CSV. Debe tener columnas 'texto' y 'clase'.

    Retorna:
        textos (list[str]): Lista de mensajes en lenguaje natural.
        clases (list[str]): Lista de etiquetas de clase correspondientes.
    """
    textos = []
    clases = []

    with open(ruta_csv, encoding="utf-8", newline="") as archivo:
        lector = csv.DictReader(archivo)
        for fila in lector:
            textos.append(fila["texto"].strip())
            clases.append(fila["clase"].strip())

    print(f"[INFO] Dataset cargado: {len(textos)} registros.")

    # Mostrar distribución de clases
    conteo_clases = {}
    for c in clases:
        conteo_clases[c] = conteo_clases.get(c, 0) + 1
    print("\nDistribución de clases:")
    for cls, n in sorted(conteo_clases.items()):
        print(f"  {cls:<25} {n} registros")

    return textos, clases


# ── EJECUTAR ───────────────────────────────────────────────────────────
textos_dataset, clases_dataset = cargar_dataset(RUTA_CSV)
```

```python
def generar_embeddings(textos, modelo_nombre="paraphrase-multilingual-MiniLM-L12-v2"):
    """
    Convierte cada texto en un vector numérico de alta dimensión (embedding).

    Por qué se hace así:
        Los modelos de lenguaje como sentence-transformers aprenden
        representaciones semánticas del texto. Textos con significado
        parecido producen vectores cercanos en el espacio vectorial.
        Esto permite que el KNN compare mensajes por su SIGNIFICADO,
        no solo por palabras literales.

        Se elige 'paraphrase-multilingual-MiniLM-L12-v2' porque:
          - Soporta español de forma nativa.
          - Es liviano y rápido (adecuado para Colab).
          - Produce embeddings de 384 dimensiones con buena calidad semántica.

        encode() retorna numpy arrays; se convierten a listas puras de
        Python para implementar los cálculos de distancia manualmente
        sin depender de numpy.

    Parámetros:
        textos (list[str])    : Lista de textos a transformar.
        modelo_nombre (str)   : Nombre del modelo de sentence-transformers.

    Retorna:
        modelo                        : El modelo cargado (reutilizable).
        embeddings (list[list[float]]): Matriz de vectores numéricos.
    """
    print(f"[INFO] Cargando modelo: {modelo_nombre} ...")
    modelo = SentenceTransformer(modelo_nombre)

    print("[INFO] Generando embeddings del dataset...")
    embeddings = [list(map(float, vec)) for vec in modelo.encode(textos, show_progress_bar=True)]
    print(f"\n[INFO]  Embeddings generados. Dimensión de cada vector: {len(embeddings[0])}")
    return modelo, embeddings


# ── EJECUTAR ───────────────────────────────────────────────────────────
modelo, embeddings_dataset = generar_embeddings(textos_dataset)
```

```python
# MÉTRICA 1: SIMILITUD DEL COSENO
# Fórmula: cos(θ) = (A · B) / (||A|| * ||B||)

def similitud_coseno(vec_a, vec_b):
    """
    Calcula la similitud del coseno entre dos vectores.

    Por qué se usa:
        Mide el ÁNGULO entre vectores, ignorando magnitud. Ideal para
        embeddings: dos textos similares apuntan en la misma dirección
        aunque sus vectores tengan longitudes distintas.
        Rango: [-1, 1]; cercano a 1 = muy similares.
    """
    producto_punto = sum(a * b for a, b in zip(vec_a, vec_b))
    norma_a = sum(a ** 2 for a in vec_a) ** 0.5
    norma_b = sum(b ** 2 for b in vec_b) ** 0.5
    if norma_a == 0 or norma_b == 0:
        return 0.0
    return producto_punto / (norma_a * norma_b)


def distancia_coseno(vec_a, vec_b):
    """
    Convierte similitud coseno en distancia para KNN.

    Por qué se hace así:
        KNN busca el MÍNIMO. Como similitud alta = mejor, invertimos:
        distancia = 1 - similitud. Rango: [0, 2]; 0 = idénticos.
    """
    return 1.0 - similitud_coseno(vec_a, vec_b)



# MÉTRICA 2: DISTANCIA L1 (MANHATTAN)
# Fórmula: L1 = Σ |a_i - b_i|

def distancia_l1(vec_a, vec_b):
    """
    Calcula la distancia L1 (Manhattan) entre dos vectores.

    Por qué se usa:
        Suma diferencias absolutas. Menos sensible a valores extremos
        (outliers) que L2 porque no eleva al cuadrado. Robusta en
        espacios de alta dimensión como los embeddings.
    """
    return sum(abs(a - b) for a, b in zip(vec_a, vec_b))



# MÉTRICA 3: DISTANCIA L2 (EUCLIDIANA)
# Fórmula: L2 = sqrt( Σ (a_i - b_i)^2 )

def distancia_l2(vec_a, vec_b):
    """
    Calcula la distancia L2 (Euclidiana) entre dos vectores.

    Por qué se usa:
        Distancia geométrica clásica. Penaliza diferencias grandes.
        Caso especial de Minkowski con p=2. Muy intuitiva pero puede
        verse afectada en espacios de alta dimensión.
    """
    return sum((a - b) ** 2 for a, b in zip(vec_a, vec_b)) ** 0.5



# MÉTRICA 4: DISTANCIA DE MINKOWSKI
# Fórmula: ( Σ |a_i - b_i|^p )^(1/p)

def distancia_minkowski(vec_a, vec_b, p=3):
    """
    Calcula la distancia de Minkowski de orden p entre dos vectores.

    Por qué se usa:
        Generalización de L1 y L2:
          p=1 → Manhattan (L1)
          p=2 → Euclidiana (L2)
          p→∞ → Chebyshev
        Con p=3 exploramos un comportamiento intermedio que puede
        capturar patrones distintos en el espacio de embeddings.
    """
    return sum(abs(a - b) ** p for a, b in zip(vec_a, vec_b)) ** (1.0 / p)


# ── Registro de métricas disponibles ──────────────────────────────────
# Cada tupla: (nombre_legible, función_de_distancia)
metricas = [
    ("Coseno",          distancia_coseno),
    ("Minkowski(p=3)",  lambda a, b: distancia_minkowski(a, b, p=3)),
    ("L1-Manhattan",    distancia_l1),
    ("L2-Euclidiana",   distancia_l2),
]

print(" Métricas definidas:", [m[0] for m in metricas])
```

```python
def calcular_distancias(vector_consulta, embeddings_dataset, metrica):
    """
    Calcula la distancia entre el vector consulta y TODOS los del dataset.

    Por qué se hace así:
        El corazón del KNN es la búsqueda exhaustiva: comparar el punto
        nuevo contra cada punto del conjunto de entrenamiento. El resultado
        es una lista (distancia, índice) lista para ordenar.
    """
    distancias = []
    for idx, vec in enumerate(embeddings_dataset):
        d = metrica(vector_consulta, vec)
        distancias.append((d, idx))
    return distancias


def seleccionar_k_vecinos(distancias, k):
    """
    Selecciona los K vecinos con menor distancia.

    Por qué se hace así:
        Ordenamos de menor a mayor distancia y tomamos los primeros K.
        Son los K puntos del dataset más similares al texto de consulta.
    """
    distancias_ordenadas = sorted(distancias, key=lambda x: x[0])
    return distancias_ordenadas[:k]


def votar_clase(vecinos, clases_dataset):
    """
    Decide la clase por votación mayoritaria entre los K vecinos.

    Por qué se hace así:
        KNN es un clasificador democrático local: la clase más frecuente
        entre los K vecinos gana. En caso de empate, se toma la clase
        con mayor conteo; si persiste el empate, orden lexicográfico.
    """
    conteo = {}
    for _, idx in vecinos:
        clase = clases_dataset[idx]
        conteo[clase] = conteo.get(clase, 0) + 1
    clase_predicha = max(conteo, key=lambda c: (conteo[c], c))
    return clase_predicha, conteo


def clasificar(vector_consulta, embeddings_dataset, clases_dataset, k, metrica):
    """
    Pipeline completo del KNN: distancias → vecinos → votación.

    Por qué se hace así:
        Encapsula los tres pasos en una función reutilizable con
        cualquier métrica y cualquier valor de K.

    Retorna:
        clase_predicha (str), vecinos (list), conteo (dict)
    """
    distancias = calcular_distancias(vector_consulta, embeddings_dataset, metrica)
    vecinos    = seleccionar_k_vecinos(distancias, k)
    clase_predicha, conteo = votar_clase(vecinos, clases_dataset)
    return clase_predicha, vecinos, conteo


print(" Algoritmo KNN implementado correctamente.")
```

```python
def mostrar_resultado(texto_consulta, clase_predicha, vecinos, conteo,
                      clases_dataset, textos_dataset, nombre_metrica, k):
    """
    Imprime los resultados de clasificación de forma clara y detallada.

    Por qué se hace así:
        El enunciado exige mostrar 'información suficiente para entender
        el proceso de decisión'. Se muestran: clase predicha, votos por
        clase y los K vecinos con su texto, clase y distancia exacta.
    """
    sep = "=" * 70
    print(f"\n{sep}")
    print(f"  CLASIFICACIÓN  |  Métrica: {nombre_metrica}  |  K={k}")
    print(sep)
    print(f"  Texto ingresado : {texto_consulta}")
    print(f"  Clase predicha  : >>>  {clase_predicha.upper()}  <<<")

    print(f"\n  Votos por clase:")
    for cls, votos in sorted(conteo.items(), key=lambda x: -x[1]):
        barra = "█" * votos
        print(f"    {cls:<25} {barra} ({votos})")

    print(f"\n  {k} vecinos más cercanos:")
    print(f"  {'#':<3} {'Distancia':<14} {'Clase':<25} Texto")
    print(f"  {'-'*3} {'-'*14} {'-'*25} {'-'*30}")
    for i, (dist, idx) in enumerate(vecinos, 1):
        texto_corto = textos_dataset[idx][:60] + ("..." if len(textos_dataset[idx]) > 60 else "")
        print(f"  {i:<3} {dist:<14.6f} {clases_dataset[idx]:<25} {texto_corto}")
    print(sep)


print(" Función de presentación lista.")
```

```python
def pruebas_automaticas(modelo, embeddings_dataset, textos_dataset,
                        clases_dataset, metricas, k=5):
    """
    Clasifica textos predefinidos con todas las métricas y muestra
    comparativamente si el sistema acierta en cada caso.

    Por qué se hace así:
        El enunciado pide 'evidencia de pruebas con varios textos de
        entrada' y 'comparación entre distintas métricas'. Esta función
        produce esa evidencia de forma reproducible para el informe.
    """
    textos_prueba = [
        ("El agua sale turbia y huele muy mal desde ayer.",             "agua_contaminada"),
        ("Se desbordó el alcantarillado frente a mi casa.",             "alcantarillado"),
        ("La presión del agua es muy baja y no llega al segundo piso.", "baja_presion"),
        ("Hay una fuga grande de agua en la tubería de la esquina.",    "fuga_agua"),
        ("Solicito reconexión del servicio de agua en mi domicilio.",   "reconexion_servicio"),
        ("Llevamos dos días completos sin servicio de agua.",           "sin_servicio"),
        # Casos ambiguos (para análisis de errores):
        ("El agua que sale tiene un color extraño y sabe raro.",        "agua_contaminada"),
        ("No tenemos agua hace varios días en el barrio.",              "sin_servicio"),
    ]

    print("\n" + "="*70)
    print("  PRUEBAS AUTOMÁTICAS  (K=" + str(k) + ")")
    print("="*70)
    print(f"  {'Texto':<52} {'Esperada':<22} ", end="")
    for nombre_m, _ in metricas:
        print(f"{nombre_m:<17}", end="")
    print()
    print("  " + "-"*120)

    for texto, clase_esperada in textos_prueba:
        vector = list(map(float, modelo.encode([texto])[0]))
        texto_corto = texto[:50] + "..." if len(texto) > 50 else texto
        print(f"  {texto_corto:<52} {clase_esperada:<22} ", end="")

        for nombre_metrica, funcion_metrica in metricas:
            clase_pred, _, _ = clasificar(
                vector, embeddings_dataset, clases_dataset, k, funcion_metrica
            )
            acierto = "✓" if clase_pred == clase_esperada else "✗"
            print(f"{clase_pred:<15} {acierto}  ", end="")
        print()

    print("\n  ✓ = Acierto    ✗ = Error")
    print("="*70)


# ── EJECUTAR ───────────────────────────────────────────────────────────
pruebas_automaticas(modelo, embeddings_dataset, textos_dataset,
                    clases_dataset, metricas, k=5)
```

```python
def evaluar_clasificador(embeddings_dataset, clases_dataset, k, metrica,
                         nombre_metrica, n_muestras=100):
    """
    Estima la exactitud con leave-one-out simplificado.

    Por qué se hace así:
        Para evaluar honestamente, cada punto de prueba se EXCLUYE del
        dataset de entrenamiento antes de clasificarlo. Así evitamos
        que el modelo 'haga trampa' encontrándose a sí mismo como
        vecino más cercano (distancia = 0).

    Retorna:
        exactitud (float): proporción de aciertos en [0, 1].
    """
    indices = list(range(len(embeddings_dataset)))
    muestra = random.sample(indices, min(n_muestras, len(indices)))

    correctos = 0
    for idx_consulta in muestra:
        emb_temp = [embeddings_dataset[i] for i in indices if i != idx_consulta]
        cls_temp = [clases_dataset[i]     for i in indices if i != idx_consulta]
        clase_pred, _, _ = clasificar(
            embeddings_dataset[idx_consulta], emb_temp, cls_temp, k, metrica
        )
        if clase_pred == clases_dataset[idx_consulta]:
            correctos += 1

    exactitud = correctos / len(muestra)
    print(f"  [{nombre_metrica:<16}]  Exactitud: {exactitud*100:.1f}%  "
          f"({correctos}/{len(muestra)} correctos)")
    return exactitud


# ── EJECUTAR ───────────────────────────────────────────────────────────
print("="*60)
print("  EVALUACIÓN DE EXACTITUD  (K=5, 100 muestras aleatorias)")
print("="*60)

random.seed(42)  # Reproducibilidad
resultados_evaluacion = {}
for nombre_metrica, funcion_metrica in metricas:
    exactitud = evaluar_clasificador(
        embeddings_dataset, clases_dataset,
        k=5, metrica=funcion_metrica,
        nombre_metrica=nombre_metrica,
        n_muestras=100
    )
    resultados_evaluacion[nombre_metrica] = exactitud
```

```python
def graficar_comparacion_metricas(resultados_evaluacion):
    """
    Gráfica de barras comparando la exactitud de cada métrica.

    Por qué se hace así:
        Una comparación visual permite apreciar de un vistazo qué métrica
        funciona mejor para este dataset. La línea base aleatoria (1/6 clases)
        sirve como referencia para cuantificar la mejora real del modelo.
    """
    nombres = list(resultados_evaluacion.keys())
    valores = [v * 100 for v in resultados_evaluacion.values()]
    colores = ["#2196F3", "#4CAF50", "#FF9800", "#E91E63"]
    linea_base = 100 / 6  # 6 clases balanceadas

    fig, ax = plt.subplots(figsize=(9, 5))
    barras = ax.bar(nombres, valores, color=colores[:len(nombres)],
                    edgecolor="black", linewidth=0.8, width=0.5)

    ax.set_title("Exactitud por métrica — KNN Acueducto Facatativá",
                 fontsize=13, fontweight="bold", pad=15)
    ax.set_ylabel("Exactitud (%)", fontsize=11)
    ax.set_ylim(0, 105)
    ax.axhline(y=linea_base, color="gray", linestyle="--", linewidth=1.5,
               label=f"Línea base aleatoria ({linea_base:.1f}%)")

    for barra, val in zip(barras, valores):
        ax.text(barra.get_x() + barra.get_width() / 2,
                barra.get_height() + 1.5,
                f"{val:.1f}%", ha="center", va="bottom",
                fontsize=12, fontweight="bold")

    ax.legend(fontsize=10)
    ax.grid(axis="y", alpha=0.3)
    plt.tight_layout()
    plt.savefig("comparacion_metricas.png", dpi=150, bbox_inches="tight")
    plt.show()
    print("[INFO] Gráfica guardada: comparacion_metricas.png")


# ── EJECUTAR ───────────────────────────────────────────────────────────
graficar_comparacion_metricas(resultados_evaluacion)
```

```python
def graficar_efecto_k(embeddings_dataset, clases_dataset, metrica,
                      nombre_metrica, valores_k=None):
    """
    Evalúa y grafica el efecto del parámetro K en la exactitud.

    Por qué se hace así:
        K es el hiperparámetro más importante en KNN:
          - K muy pequeño → sobreajuste (sensible al ruido)
          - K muy grande  → subajuste (pierde detalle local)
        Esta curva permite identificar el K óptimo para este dataset.
    """
    if valores_k is None:
        valores_k = [1, 3, 5, 7, 9, 11, 15, 20]

    print(f"[INFO] Analizando efecto de K con métrica: {nombre_metrica}")
    random.seed(42)
    exactitudes = []

    for k in valores_k:
        indices = list(range(len(embeddings_dataset)))
        muestra = random.sample(indices, 80)
        correctos = 0
        for idx_consulta in muestra:
            emb_temp = [embeddings_dataset[i] for i in indices if i != idx_consulta]
            cls_temp = [clases_dataset[i]     for i in indices if i != idx_consulta]
            clase_pred, _, _ = clasificar(
                embeddings_dataset[idx_consulta], emb_temp, cls_temp, k, metrica
            )
            if clase_pred == clases_dataset[idx_consulta]:
                correctos += 1
        exactitud = correctos / len(muestra)
        exactitudes.append(exactitud * 100)
        print(f"    K={k:>2} → {exactitud*100:.1f}%")

    fig, ax = plt.subplots(figsize=(9, 5))
    ax.plot(valores_k, exactitudes, marker="o", color="#2196F3",
            linewidth=2.5, markersize=8, markerfacecolor="white",
            markeredgewidth=2.5)
    ax.fill_between(valores_k, exactitudes, alpha=0.12, color="#2196F3")

    # Marcar el K óptimo
    k_opt = valores_k[exactitudes.index(max(exactitudes))]
    ax.axvline(x=k_opt, color="red", linestyle=":", linewidth=1.5,
               label=f"K óptimo = {k_opt}")

    ax.set_title(f"Efecto del parámetro K en la exactitud\n(Métrica: {nombre_metrica})",
                 fontsize=13, fontweight="bold")
    ax.set_xlabel("Valor de K", fontsize=11)
    ax.set_ylabel("Exactitud (%)", fontsize=11)
    ax.set_xticks(valores_k)
    ax.set_ylim(0, 105)
    ax.grid(True, alpha=0.35)
    ax.legend(fontsize=10)
    plt.tight_layout()
    plt.savefig("efecto_k.png", dpi=150, bbox_inches="tight")
    plt.show()
    print(f"[INFO] K óptimo encontrado: {k_opt}")
    print("[INFO] Gráfica guardada: efecto_k.png")


# ── EJECUTAR ───────────────────────────────────────────────────────────
# Analizamos con coseno (mejor métrica para embeddings)
graficar_efecto_k(
    embeddings_dataset, clases_dataset,
    distancia_coseno, "Coseno",
    valores_k=[1, 3, 5, 7, 9, 11, 15, 20]
)
```

```python
def modo_interactivo(modelo, embeddings_dataset, textos_dataset,
                     clases_dataset, metricas, k_default=5):
    """
    Permite al usuario ingresar mensajes por teclado para clasificarlos.

    Por qué se hace así:
        El enunciado exige que el programa permita que el usuario escriba
        un nuevo mensaje por teclado. Este bucle interactivo solicita el
        texto, permite cambiar K en tiempo real, y clasifica con las 4
        métricas simultáneamente para facilitar la comparación.

    Comandos:
        'salir'  → termina el bucle
        'k=N'    → cambia el valor de K a N
        (cualquier otro texto) → se clasifica
    """
    k = k_default
    print("\n" + "="*70)
    print("  MODO INTERACTIVO — Clasificador KNN Acueducto Facatativá")
    print("  Comandos: 'salir' para terminar | 'k=N' para cambiar K")
    print("="*70)

    while True:
        texto = input(f"\n  [K={k}] Ingresa tu mensaje: ").strip()

        if texto.lower() == "salir":
            print("  Sesión terminada.")
            break

        if texto.lower().startswith("k="):
            try:
                k = int(texto[2:])
                print(f"   K actualizado a {k}")
            except ValueError:
                print("   Valor de K inválido.")
            continue

        if not texto:
            print("    El texto no puede estar vacío.")
            continue

        vector = list(map(float, modelo.encode([texto])[0]))

        print(f"\n  Resultados con K={k}:")
        print(f"  {'Métrica':<18} {'Clase predicha':<25} Votos")
        print(f"  {'-'*18} {'-'*25} {'-'*20}")
        for nombre_metrica, funcion_metrica in metricas:
            clase_pred, vecinos, conteo = clasificar(
                vector, embeddings_dataset, clases_dataset, k, funcion_metrica
            )
            votos_str = " | ".join(f"{c}:{v}" for c, v in
                                    sorted(conteo.items(), key=lambda x: -x[1]))
            print(f"  {nombre_metrica:<18} {clase_pred:<25} {votos_str}")

        # Mostrar detalle completo solo de la primera métrica
        nombre_m0, func_m0 = metricas[0]
        clase_pred0, vecinos0, conteo0 = clasificar(
            vector, embeddings_dataset, clases_dataset, k, func_m0
        )
        print(f"\n  [Detalle de vecinos — {nombre_m0}]")
        print(f"  {'#':<3} {'Dist':<10} {'Clase':<25} Texto")
        for i, (dist, idx) in enumerate(vecinos0, 1):
            tc = textos_dataset[idx][:55] + "..." if len(textos_dataset[idx]) > 55 else textos_dataset[idx]
            print(f"  {i:<3} {dist:<10.5f} {clases_dataset[idx]:<25} {tc}")


# ── EJECUTAR ───────────────────────────────────────────────────────────
modo_interactivo(modelo, embeddings_dataset, textos_dataset,
                 clases_dataset, metricas, k_default=5)
```
